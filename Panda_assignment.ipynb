```python
import pandas as pd
from pathlib import Path
```

```python
school_data_to_load = Path("PyCitySchools/Resources/schools_complete.csv")
student_data_to_load = Path("PyCitySchools/Resources/students_complete.csv")

school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
total_unique_schools = len(school_data["school_name"].unique())
total_unique_schools
```

```python
total_students = len(student_data["student_name"])
total_students
```

```python
total_budget = school_data["budget"].sum()
total_budget
```

```python
school_data_complete.columns
```

```python
# avg math score
avg_math_score = (student_data["maths_score"].mean())
avg_math_score
```

```python
# avg reading score
avg_reading_score = (student_data["reading_score"].mean())
avg_reading_score
```

```python
cond1_percent_math = student_data[student_data["maths_score"] >= 50]
cond2_percent_math = len(cond1_percent_math["maths_score"])
pass_math_percent = ((cond2_percent_math)/total_students)*100
pass_math_percent
```

```python
cond1_percent_reading = student_data[student_data["reading_score"] >= 50]
cond2_percent_reading = (len(cond1_percent_reading["reading_score"]))
pass_reading_percent = ((cond2_percent_reading)/total_students)*100
pass_reading_percent
```

```python
# cond2_percentage = student_data[student_data["reading_score"]>=50] & [student_data["maths_score"] >= 50]

cond1_percentage = student_data[(student_data["reading_score"] >= 50) & (student_data["maths_score"] >= 50)]
cond2_percentage = (len(cond1_percentage))
overall_percent = ((cond2_percentage)/total_students)*100
overall_percent
```

```python
Score_display =[{"Total School":total_unique_schools,
                 "Total Students":total_students,
                 "Total Budget":total_budget,
                 "Average Math score":avg_math_score,
                 "Average Reading score":avg_reading_score,
                 "% Passing Math":pass_math_percent,
                 "% Passing Reading":pass_reading_percent,
                 "% Overall Passing":overall_percent}]

Score_display_df = pd.DataFrame(Score_display)
Score_display_df["Total Budget"] = Score_display_df["Total Budget"].map('${:,.2f}'.format)
Score_display_df
```

```python
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
school_summary['Per Student Budget'] = school_summary['Per Student Budget'].astype(float)
```

```python
school_summary["Spending Ranges (Per Student)"] = pd.cut(x=school_summary["Per Student Budget"], 
                                                         bins = spending_bins, 
                                                         labels=labels)
```

```python
spending_math_scores = school_summary.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_summary.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_summary.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_summary.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_summary.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
summarised_spending = pd.DataFrame({"Average Math Score": spending_math_scores,
                              "Average Reading Score": spending_reading_scores,
                              "% Passing Math": spending_passing_math,
                              "% Passing Reading":spending_passing_reading,
                              "% Overall Passing":overall_passing_spending})
```
